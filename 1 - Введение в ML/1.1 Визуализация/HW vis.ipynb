```python
price = train.pivot_table(
            index='OverallCond',
            columns='HouseStyle',
            values='SalePrice').fillna(0).applymap(float)
price
```

Построена heatmap Для оценки зависимости средней цены на пересечении двух категориальных признаков: типа дома и состояния материалов. Видно, что в целом на стоимость сильно влияет состояние материалов. Также выделяются 2,5 этажные "качественные" дома.

```python
sns.heatmap(price, annot=True, fmt=".0f", 
            linewidths=.5);
```

Построет barchart зависиомости суммарной стоимости проданных домов от класса здания. Видно, что больше всего оборота было по одно- и двухэтажным домам 1946 года постройки и позднее. 

```python
train['NewPrice'] = train.SalePrice*0.001
train[['NewPrice','MSSubClass']]\
    .groupby('MSSubClass').sum()\
    .plot(kind='bar', rot=45);
```

```python
from plotly.offline import init_notebook_mode, iplot
import plotly
import plotly.graph_objs as go
```

```python
df = train[['ExterQual', 'BsmtQual', 'MSSubClass', 'SalePrice']]
df = df.replace(to_replace=['Gd', 'TA', 'Ex', 'Fa' ], value=[2, 1, 3, 0])
df.BsmtQual = df.BsmtQual.fillna(0)
```

```python
init_notebook_mode(connected=True)
```

```python
v_df = df.groupby('MSSubClass')[['ExterQual', 'BsmtQual']].mean()\
    .join(df.groupby('MSSubClass')[['SalePrice']].sum())
    
v_df.head()
```

С помощью модуля plotly построен scatterplot с дополнительной информацией. По осям расположены качество фундамента и качество фасада. Точками обозначены различные классы здания. Площадь кругов обозначает суммарные продажи по данному классу.

```python
trace0 = go.Scatter(
    x=v_df.ExterQual,
    y=v_df.BsmtQual,
    mode = 'markers+text',
    text = v_df.index,
    #textposition='bottom'
    marker = dict(
        size = 1/1000000*v_df.SalePrice,
        color = [
            'aqua', 'azure', 'beige', 'lightgreen',
            'lavender', 'lightblue', 'pink', 'salmon',
            'wheat', 'ivory', 'silver', 'plum'
        ]
    )
)

data = [trace0]
layout = {'title': 'Statistics of house sales'}

fig = go.Figure(data=data, layout=layout)

iplot(fig, show_link=False)
```
